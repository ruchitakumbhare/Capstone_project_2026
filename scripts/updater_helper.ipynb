```python
import numpy as np
import pandas as pd
```

Function-1 [2-Dimensional]

```python
#function1_inputs = np.load('../data/initial_data/function_1/initial_inputs.npy')
#function1_outputs = np.load('../data/initial_data/function_1/initial_outputs.npy')
function1_inputs = np.load('../data/week9/function_1/inputs.npy')
function1_outputs = np.load('../data/week9/function_1/outputs.npy')

# ==== Import New Data ====

f1_new_inputs = np.array([0.446826,0.999040])
f1_new_output = np.array([np.float64(-3.2630040678240334e-119)])


# ==== Join old and new data ====

function1_inputs_updated = np.vstack([function1_inputs, f1_new_inputs])
print(function1_inputs_updated)

function1_outputs_updated = np.concatenate((function1_outputs, f1_new_output))
print(function1_outputs_updated)


# ==== Export Updated Data Sets as .npy files ====

# Make sure to change file name each week so old versions aren't overwritten.
np.save('../data/week10/function_1/inputs', function1_inputs_updated)
np.save('../data/week10/function_1/outputs', function1_outputs_updated)
```
